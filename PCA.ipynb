# PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
wine=pd.read_csv('wine.csv')
```

```python
wine
```

```python
wine.describe()
```

```python
wine1=wine.iloc[:,1:]
```

```python
wine1
```

```python
wine1.info()
```

```python
WINE=wine1.values
WINE
```

```python
wine2=scale(WINE)
```

```python
wine2
```

```python
pca=PCA()
pca_val=pca.fit_transform(wine2)
pca_val
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
```

```python
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
```

```python
var1
```

```python
plt.plot(var1,color='green')
```

```python
pca_val[:,0:3]
```

```python
fig=plt.figure(figsize=(10,7))
x=pca_val[:,0:1]
y=pca_val[:,1:2]
z=pca_val[:,2:3]
plt.scatter(x,y,z)
```

```python
y=pd.DataFrame(pca_val[:,0:3],columns=['PC1','PC2','PC3'])
```

```python
y
```

```python
df=pd.concat([wine[['Type']],y],axis=1)
```

```python
df
```

```python
sns.scatterplot(data=df,x='PC1',y='PC2',hue='Type')
```

```python
sns.scatterplot(data=df,x='PC1',y='PC3',hue='Type')
```

# KMeans

```python
wine2
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(wine2)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
```

```python
ml=KMeans(n_clusters=5)
ml.fit(wine2)
```

```python
ml.labels_
```

```python
md=pd.DataFrame(ml.labels_)
```

```python
md
```

```python
pd.concat([pa,md],axis=1)
```
